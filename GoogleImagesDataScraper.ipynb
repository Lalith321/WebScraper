```python
import requests
from bs4 import BeautifulSoup
import base64
from urllib.request import urlopen
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def create_directory(keyword):
    if not os.path.exists(keyword):
        os.makedirs(keyword)
```

```python
def fetch_image_urls(keyword, num_images):
    options = webdriver.ChromeOptions()
    options.add_argument("--headless-new")
    driver = webdriver.Chrome(options=options)
    url = f"https://www.google.com/search?hl=en&tbm=isch&q={keyword}"
    # url = f"https://www.pinterest.com/search/pins/?q={keyword}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    
    driver.get(url)
    
    # Search for images
    search_box = driver.find_element(By.NAME, 'q')
    search_query = ' high resolution'
    search_box.send_keys(search_query)
    search_box.send_keys(Keys.RETURN)
    image_urls = []
    image_index = 0
    while len(image_urls) < num_images:
        image_index = image_index+1
        try:
            # Wait for images to load
            time.sleep(3)

            # Find and click on the first image
            first_image = driver.find_element(By.XPATH, f'/html/body/div[4]/div/div[15]/div/div[2]/div[2]/div/div/div/div/div[1]/div/div/div[{image_index}]/div[2]/h3/a/div/div/div/g-img/img')
            first_image.click()
            time.sleep(5)

            actual_image = driver.find_element(By.XPATH, '//*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[3]/div[1]/a/img[1]')
            image_url = actual_image.get_attribute('src')
            image_urls.append(image_url)
            
            with open(f"{keyword}.txt", "ab") as file:
                file.write(image_url.encode()+b'\n')

            # print(image_url)
        except Exception as e1:
            print("exception1", e1)
    return image_urls
```

```python
def download_images(image_urls, category):
    for i, url in enumerate(image_urls):
        # print(url)
        try:
            response = requests.get(url)
            with open(f"{category}/{category}_{i+1}.jpg", "wb") as file:
                file.write(response.content)
        except Exception as e2:
            print("exception2", e2)
```

```python
def main(categories, num_images):
    for category in categories:
        create_directory(category)
        image_urls = fetch_image_urls(category, num_images)
        download_images(image_urls, category)
        print(f"Downloaded {len(image_urls)} images of {category}")
```

```python
if __name__ == "__main__":
    categories = ['plate', 'bowls', 'dustbin', 'sofa', 'cup and saucer', 'fork', 'glass', 'salt and pepper shakers', 'Condiment dispensers', 'serving trays', 'tray stand']
    num_images = 600
    main(categories, num_images)
```
